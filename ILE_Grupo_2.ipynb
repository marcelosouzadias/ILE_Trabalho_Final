```python
#Importação das Bibliotecas
#Verificar como importar a biblioteca pandas
from random import randrange

#função do menu do panda
def criaMenu():    
    print("|------------------------------------------------------------------------------------|")
    print("|------------------Introdução Linguagem Estatistica - Grupo 2------------------------|")
    print("|-----------------Marcelo | Lorena | Renato Dalmo | Vitor Cunha----------------------|")
    print("|------------------------------------------------------------------------------------|")    
    print()
    
criaMenu()
#listagem de estados do Brasil
listMenu = ["AC", "AL", "AM", "AP", "BA", "CE", "DF", "ES", "GO", "MA", "MG", "MS", "MT", "PA", "PB", "PE", "PI", "PR", "RJ", "RN", "RO", "RR", "RS", "SC", "SE", "SP", "TO"]

print(listMenu)

#Rodar o sistema enquanto colocar o xx
while True:
    vEstado = input('Informe uma sigla do Estado ou XX para sair!').upper()
    achou = 0
    #Verifica se quer sair do sistema
    if vEstado == 'XX':
        print()
        print()
        for item in listMenu:
            print('Estado Informado:', item)
            print('Quantidade de Vendas:', randrange(1,500))
            print('Receita total:', str(randrange(1,1500)) + ',00')
            print()
        print("Saindo do sistema!")
        break
    else:
        #Percorre a lista dos estados
        for item in listMenu:
            if item == vEstado:
                achou += 1
                criaMenu()
                print()
                print('Estado Informado:', vEstado)
                print('Quantidade de Vendas:', randrange(1,500))
                print('Receita total:', str(randrange(1,1500)) + ',00')
                print()
                print()
                break  
        if achou == 0:
            print('Estado não informado corretamente!')
```
