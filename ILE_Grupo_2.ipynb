```python
import pandas as pd
import sqlite3 as cn

#FUNÇÃO QUE RETORNA A LISTA DOS ESTADOS VENDIDOS
def getUF():
    #conexão com o banco de dados
    conexao = cn.connect('exercício_final.db')
    
    tb_Clientes = pd.read_sql("select id_cliente, des_uf from tb_clientes", conexao)
    
    #tabela de pedidos
    tb_pedidos = pd.read_csv('pedidos.csv')
    conexao.close()
    
    #join cliente x pedidos    
    join_cliente_pedido = pd.merge(tb_Clientes, tb_pedidos, how='inner', on=['id_cliente','id_cliente']) 
    
    #filtrar somente id_cliente, des_uf
    cliente_pedido = join_cliente_pedido[['des_uf']]  
    tb_uf = cliente_pedido.groupby('des_uf').count()  
    ufs = []
    for index, row in tb_uf.iterrows():
        ufs.append(index)
    return ufs
```

```python
import pandas as pd
import sqlite3 as cn

#FUNÇÃO QUE RETORNA O VALOR VENDAS
def getUF_Valor_Vendas(estado):
    #conexão com o banco de dados
    conexao = cn.connect('exercício_final.db')

    #tabela de Clientes
    tb_Clientes = pd.read_sql("select id_cliente, des_uf from tb_clientes where des_uf = '" + str(estado).upper() + "'", conexao)    

    #Tabela de Produtos
    tb_Produtos = pd.read_sql("select * from tb_produtos", conexao)

    #tabela de pedidos
    tb_pedidos = pd.read_csv('pedidos.csv')

    #Fecha conexão com o banco de dados
    conexao.close()

    #join cliente x pedidos    
    join_cliente_pedido = pd.merge(tb_Clientes, tb_pedidos, how='inner', on=['id_cliente','id_cliente'])
    
    #join clinte x pedidos x produtos
    join_cli_ped_prod = pd.merge(join_cliente_pedido, tb_Produtos, how='inner', on=['id_produto', 'id_produto'])
    
    #Filtro campos necessários
    tb_Geral = join_cli_ped_prod[['id_cliente','des_uf','id_produto','qtd_vendida','val_produto']]

    tb_Geral['vl_total'] = tb_Geral['qtd_vendida'] * tb_Geral['val_produto']
        
    tb_groupby = tb_Geral.groupby('des_uf').sum() 
    
    return tb_groupby.iloc[0][4]
```

```python
import pandas as pd
import sqlite3 as cn

#FUNÇÃO QUE RETORNA A QUANTIDADE DE VENDAS POR ESTADO
def getUF_qtd_Produtos(estado):
    #conexão com o banco de dados
    conexao = cn.connect('exercício_final.db')

    #tabela de Clientes
    tb_Clientes = pd.read_sql("select id_cliente, des_uf from tb_clientes where des_uf = '" + str(estado).upper() + "'", conexao)    

    #Tabela de Produtos
    tb_Produtos = pd.read_sql("select * from tb_produtos", conexao)

    #tabela de pedidos
    tb_pedidos = pd.read_csv('pedidos.csv')

    #Fecha conexão com o banco de dados
    conexao.close()

    #join cliente x pedidos    
    join_cliente_pedido = pd.merge(tb_Clientes, tb_pedidos, how='inner', on=['id_cliente','id_cliente'])

    #join clinte x pedidos x produtos
    join_cli_ped_prod = pd.merge(join_cliente_pedido, tb_Produtos, how='inner', on=['id_produto', 'id_produto'])

    #Filtro campos necessários
    tb_Geral = join_cli_ped_prod[['id_cliente','des_uf','id_produto','qtd_vendida','val_produto']]


    #tb_Geral['vl_total'] = tb_Geral['qtd_vendida'] * tb_Geral['val_produto']
    
    tb_groupby = tb_Geral.groupby('des_uf').sum() 
    
    return tb_groupby.iloc[0][2]
```

```python
import pandas as pd
import sqlite3 as cn

#FUNÇÃO QUE RETORNA A QUANTIDADE DE VENDAS POR ESTADO
def getUF_qtd_Vendas(estado):
    #conexão com o banco de dados
    conexao = cn.connect('exercício_final.db')

    #tabela de Clientes
    tb_Clientes = pd.read_sql("select id_cliente, des_uf from tb_clientes where des_uf = '" + str(estado).upper() + "'", conexao)    
   
    #tabela de pedidos
    tb_pedidos = pd.read_csv('pedidos.csv')

    #Fecha conexão com o banco de dados
    conexao.close()

    #join cliente x pedidos    
    join_cliente_pedido = pd.merge(tb_Clientes, tb_pedidos, how='inner', on=['id_cliente','id_cliente'])

    #Filtro campos necessários
    tb_Geral = join_cliente_pedido[['id_cliente','des_uf']]

    tb_Geral['qtd'] = 1
    
    tb_groupby = tb_Geral.groupby('des_uf').count() 
    
    return tb_groupby.iloc[0][1]

```
