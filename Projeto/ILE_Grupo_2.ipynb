```python
import pandas as pd
import sqlite3 as cn

#FUNÇÃO QUER RETORNA A LISTA DOS ESTADOS VENDIDOS
def getUF():
    #conexão com o banco de dados
    conexao = cn.connect('exercício_final.db')
    
    tb_Clientes = pd.read_sql("select id_cliente, des_uf from tb_clientes", conexao)
    
    #tabela de pedidos
    tb_pedidos = pd.read_csv('pedidos.csv')
    conexao.close()
    
    #join cliente x pedidos    
    join_cliente_pedido = pd.merge(tb_Clientes, tb_pedidos, how='inner', on=['id_cliente','id_cliente']) 
    
    #filtrar somente id_cliente, des_uf
    cliente_pedido = join_cliente_pedido[['des_uf']]  
    tb_uf = cliente_pedido.groupby('des_uf').count()  
    ufs = []
    for index, row in tb_uf.iterrows():
        ufs.append(index)
    return ufs

#FUNÇÃO DO RELATORIO POR ESTADO
# Função que recebe o Estado e Retorna lista com os dados:
#  - Quantidade de vendas por UF
#  - Quantidade de produtos vendidos UF
#  - Receita total para a UF
def getVendasEstado(estado):
    #conexão com o banco de dados
    conexao = cn.connect('exercício_final.db')
    
    #tabela de Clientes
    tb_Clientes = pd.read_sql("select id_cliente, des_uf from tb_clientes where des_uf = '" + str(estado).upper() + "'", conexao)    
    #tb_Clientes = pd.read_sql("select id_cliente, des_uf from tb_clientes", conexao)
     
    #tabela de Produtos
    tb_Produtos = pd.read_sql("select id_produto, val_produto from tb_produtos", conexao)
    conexao.close()
     
    #tabela de pedidos
    tb_pedidos = pd.read_csv('pedidos.csv')
    
    #join cliente x pedidos    
    join_cliente_pedido = pd.merge(tb_Clientes, tb_pedidos, how='inner', on=['id_cliente','id_cliente'])
    
    #join cliente_pedido x produto
    join_cli_ped_prod = pd.merge(join_cliente_pedido, tb_Produtos, how='inner', on=['id_produto', 'id_produto'])
    
    #filtrar somente id_cliente, des_uf
    cliente_pedido = join_cliente_pedido[['id_cliente', 'des_uf']]  
    tb_count_pedidos = cliente_pedido.groupby('des_uf').count()   
    
    #filtra somente val_produto, des_uf
    cliente_produto = join_cli_ped_prod[['des_uf', 'val_produto']]
    tb_count_produto = cliente_produto.groupby('des_uf').sum()
    
    ret_estado = [tb_count_pedidos.iloc[0][0], tb_count_produto.iloc[0][0]]
    
    return ret_estado
```
